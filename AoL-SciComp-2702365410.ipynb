```python
# Problem 3
import numpy as np
from datetime import datetime, timedelta

# Exponential parameters
a = 1966.957
b = 0.015
max_capacity = 25000

# Function to convert months to years and months
def month_to_date(start_date, months):
    start_date = datetime.strptime(start_date, "%Y-%m-%d")
    end_date = start_date + timedelta(days=months*30)  # Approximate 30 days per month
    return end_date.strftime("%B %Y")

# Set start date 
start_date = "2018-01-01"

# Find period x where bag production exceeds 25000
x_exceed = np.log(max_capacity / a) / b

# Subtract 13 months to start new warehouse construction
start_building = x_exceed - 13

# Convert months to years and months
exceed_date = month_to_date(start_date, int(np.ceil(x_exceed)))  # Round up to nearest whole month
start_building_date = month_to_date(start_date, int(np.ceil(start_building)))  # Round up to nearest whole month

print(f"Period where production exceeds 25000 bags: {x_exceed:.2f} months ({exceed_date})")
print(f"Time needed to construct new warehouse: {start_building:.2f} months ({start_building_date})")
```
